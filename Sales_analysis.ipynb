```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode-escape")
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtype
```

```python
ax=sns.countplot(data=df,x='Gender')
for a in ax.containers:
    ax.bar_label(a)
plt.show()
```

```python
total=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.title("Relation between gender and amount")
sns.barplot(x='Gender',y='Amount',data=total)
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for a in ax.containers:
    ax.bar_label(a)
```

```python
gd=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
plt.title("Relation between the age Group and Amount")
sns.barplot(x='Age Group',y='Amount',data=gd)
```

from the above graphs we can see that most of the buyers are of age group between 26-35 years female

```python
df.columns
```

```python
state_sales=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(x='State',y='Orders',data=state_sales)
```

```python
state_sales=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=state_sales)
```

From the above graph we can see that most of the orders and total sales/amount are from Up,Karnataka,Maharastra 

```python
ab=sns.countplot(data=df,x="Marital_Status",hue='Gender')
for a in ab.containers:
    ab.bar_label(a)
```

```python
gd=df.groupby(["Marital_Status"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Marital_Status',y="Amount",data=gd)
```

```python
from the above graph we can see that most of the buyers are married women 
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x="Occupation",data=df)
for a in ax.containers:
    ax.bar_label(a)
```

```python
gd=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Occupation',y="Amount",data=gd)
```

```python
from the above graph we can see that most of the buyers are working in IT,Healthcare and Aviation
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ab=sns.countplot(data=df,x="Product_Category")
for a in ab.containers:
    ab.bar_label(a)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
gd=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x='Product_Category',y="Amount",data=gd)
```

from the above the graph we can see that most sold product are from food,Cloting and electronic gadget
